```python
from tqdm import tqdm
import ll
from os import path
from skimage.color import rgb2gray
from skimage.transform import resize
from skimage.io import imread
from scipy.fftpack import dct
import numpy as np
```

```python
def phash(raw):
    img = rgb2gray(raw)
    #img = resize(img, (32, 32), anti_aliasing=True, mode='reflect')
    imgdct = dct(dct(img.T, norm='ortho').T, norm='ortho')
    imgdct = imgdct[0:8,0:8]
    avg = imgdct[1:7,1:7].mean()
    bits = np.ravel(imgdct > avg) * 1
    y = 0
    for bit in bits:
        y = (y << 1) | bit
    return y
```

```python
hashes = []
for i in tqdm(range(1,4600)):
    ipath = path.join("video/frames/frame" + format(i,"04") + ".png")
    img = imread(ipath)
    hash = ll.dhash(img)
    hashes.append(hash)
```

```python
def hamming(a, b):
    return bin(a^b).count('1')
```

```python
for i in range(1, len(hashes)):
    #print(i, ": ", bin(hashes[i]))
    print(i, hamming(hashes[i], hashes[i-1]))
```
